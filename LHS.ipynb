```python
bounds_nparr = np.array(list(bounds.values()))

sampling = LHS(xlimits=bounds_nparr, random_state=42) #генератор сэмплирования
```

```python
num = 1e2 # Количество сэмплов, которые следует нагенерировать.
sampled_data = sampling(int(num)) # стартуем генерацию
```

```python
sys.getsizeof(sampled_data)/10**6 # Размер занимаемой данными памяти в Mb.
```

```python
df_samples = pd.DataFrame(sampled_data, columns = list(bounds.keys()))
```
